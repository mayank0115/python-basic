7.what is the difference between "==" and "is" operators in python?

-"==" (Equality Operator)
Checks if the values of two objects are equal.

It's about what the objects contain, not whether they are the same object.

example
a = [1, 2, 3]
b = [1, 2, 3]
print(a == b)  # True (because contents are equal)
is (Identity Operator)
Checks if two variables point to the same object in memory.

It's about whether both names refer to the exact same object.

a = [1, 2, 3]
b = [1, 2, 3]
print(a is b)  # False (they are two different objects in memory)

12.how does the elif statement  work?

-The elif statement in Python stands for "else if", and it lets you check multiple conditions in an if-elif-else structure.

It works like this:

Python checks the if condition.

If it’s True, that block runs, and the rest (including any elif or else) is skipped.

If the if is False, Python checks the first elif.

This continues top to bottom until a condition is True.

If none of the if or elif conditions are True, the else block runs (if present).

13.what is the difference between for and while loops?

Ideal when you know the number of iterations or have a collection to loop through.
while loop – Used when you don’t know how many times to loop
